```python
cmd = r"""
def CMD(cmd):
    p = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, shell=True, encoding='utf-8')
    stack = ""
    while True:
        word = p.stdout.readline(1)
        stack += word
        if "\n" in stack:
            if "speed" in stack:
                print(stack)
            stack = ""
        if not word: break
    print(end="\n")
    return 0
"""

with open("youtube.py", "r", encoding="utf-8") as r: r = cmd + r.read().replace("os.system", "CMD")
with open("youtube.py", "w", encoding="utf-8") as w: w.write(r)
```
